# 홈앤리빙 분석 요약

이 깃허브용 노트북은 검토한 EDA 및 프로젝트 노트북의 핵심 내용을 요약합니다. 기존 분석 의도는 유지하고, 저장소 기준 상대 경로로 데이터를 불러오도록 정리했습니다.


## 범위

- 데이터셋: Olist 이커머스 CSV 파일
- 초점: 홈앤리빙 카테고리, 셀러 성장, SKU 폭, 고객 반복 구매 행동, 리텐션 전략
- 원본 노트북: `olist_home_living_eda.ipynb`, `olist_home_living_eda copy.ipynb`, `olist_home_living_project.ipynb`, `olist_home_living_project2.ipynb`


```python
from pathlib import Path

import pandas as pd

BASE_DIR = Path("..") if (Path("..") / "data").exists() else Path(".")
DATA_DIR = BASE_DIR / "data"

def load_csv(filename: str) -> pd.DataFrame:
    path = DATA_DIR / filename
    if not path.exists():
        raise FileNotFoundError(f"데이터 파일을 찾을 수 없습니다: {path}")
    return pd.read_csv(path)

orders = load_csv("olist_orders_dataset.csv")
customers = load_csv("olist_customers_dataset.csv")
order_items = load_csv("olist_order_items_dataset.csv")
payments = load_csv("olist_order_payments_dataset.csv")
reviews = load_csv("olist_order_reviews_dataset.csv")
products = load_csv("olist_products_dataset.csv")
sellers = load_csv("olist_sellers_dataset.csv")
category_translation = load_csv("product_category_name_translation.csv")

for name, df in {
    "orders": orders,
    "customers": customers,
    "order_items": order_items,
    "payments": payments,
    "reviews": reviews,
    "products": products,
    "sellers": sellers,
    "category_translation": category_translation,
}.items():
    print(name, df.shape)
```

## 재사용 가능한 병합 패턴

원본 노트북은 주문, 상품, 고객, 셀러, 결제, 번역된 카테고리명을 반복적으로 결합합니다. 아래의 간단한 병합 코드는 카테고리와 셀러 분석에 사용할 기본 테이블을 만듭니다.


```python
order_payment = payments.groupby("order_id", as_index=False)["payment_value"].sum()

items = (
    order_items
    .merge(orders[["order_id", "customer_id", "order_status", "order_purchase_timestamp"]], on="order_id", how="left")
    .merge(customers[["customer_id", "customer_unique_id", "customer_state"]], on="customer_id", how="left")
    .merge(products[["product_id", "product_category_name"]], on="product_id", how="left")
    .merge(category_translation, on="product_category_name", how="left")
    .merge(sellers[["seller_id", "seller_state"]], on="seller_id", how="left")
    .merge(order_payment, on="order_id", how="left")
)

items["product_category_name_english"] = items["product_category_name_english"].fillna("unknown")
items.head()
```

## 홈앤리빙 카테고리 묶음

검토한 노트북은 아래와 같은 홈앤리빙 카테고리를 강조합니다.

- `housewares`
- `furniture_decor`
- `bed_bath_table`


```python
home_living_categories = ["housewares", "furniture_decor", "bed_bath_table"]

home_living = items[items["product_category_name_english"].isin(home_living_categories)].copy()

category_perf = (
    home_living.groupby("product_category_name_english")
    .agg(
        order_count=("order_id", "nunique"),
        seller_count=("seller_id", "nunique"),
        customer_count=("customer_unique_id", "nunique"),
        item_revenue=("price", "sum"),
    )
    .sort_values("item_revenue", ascending=False)
)

category_perf
```

## 셀러 성장 관점

프로젝트 전략은 셀러의 SKU 폭에 초점을 둡니다. 원본 메모에서 강조한 18개 SKU 목표에 가까운 10~17개 SKU 셀러를 실무적으로 중요한 세그먼트로 볼 수 있습니다.


```python
seller_summary = (
    home_living.groupby("seller_id")
    .agg(
        sku_count=("product_id", "nunique"),
        order_count=("order_id", "nunique"),
        revenue=("price", "sum"),
    )
    .sort_values("revenue", ascending=False)
)

jump_up_candidates = seller_summary.query("10 <= sku_count < 18")

seller_summary.head(), jump_up_candidates.head()
```

## 전략 요약

- 이미 판매 성과가 있는 셀러를 우선합니다.
- 목표에 가까운 셀러가 18개 이상 SKU로 확장하도록 돕습니다.
- 셀러 상품 구색 확장과 3회 이상 구매 고객 대상 리텐션 캠페인을 연결합니다.
- 셀러 SKU 수, 매출 집중도, 주문 수, 반복 구매 행동을 추적합니다.
